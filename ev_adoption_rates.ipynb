### Clustering Analysis

Finding the optimal number of clusters with the Elbow Method

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

#Selecting relevant variables for clustering
selected_features = ['Charging Station Density', 'EV Adoption Rate (%)', 'Average EV Age']
demo_selected = final_demographic[selected_features]

#Standardizing the data
scaler = StandardScaler()
demo_scaled = scaler.fit_transform(demo_selected)

#Elbow Method to find optimal k
wcss = []
k_range = range(1, 11)  

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(demo_scaled)
    wcss.append(kmeans.inertia_)  #Storing WCSS values

#Plotting the Elbow Method graph
plt.figure(figsize=(8, 5))
plt.plot(k_range, wcss, marker="o", linestyle="--")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("WCSS (Within-Cluster Sum of Squares)")
plt.title("Elbow Method for Optimal k Selection")
plt.xticks(k_range)
plt.grid(False)
plt.show()

#Printing WCSS values
for k, value in zip(k_range, wcss):
    print(f"k={k}, WCSS={value:.2f}")
```

Based on the Elbow method, the optimal number of clusters is k = 4 because after that, WCSS reduction starts slowing down substantilly

#### Running K-Means (K=4)

```python
from sklearn.cluster import KMeans

#Running K-Means with k=4
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
demo_selected['Cluster'] = kmeans.fit_predict(demo_scaled)

#Adding cluster labels to the original dataframe
final_demographic['Cluster'] = demo_selected['Cluster']

final_demographic.head()
```

```python
final_demographic['Cluster'].value_counts()
```

It seems like San Juan county is the only data point on cluster 2, meaning that it could potentially be an outlier

Charging station density is high compared to other clusters. However, EV adoption rate remains low

```python
#Selecting relevant variables for clustering
selected_features_wo = ['Charging Station Density', 'EV Adoption Rate (%)', 'Average EV Age']
demo_selected_wo = final_demographic_wo_sj[selected_features_wo]

#Standardizing the data
scaler_wo = StandardScaler()
demo_scaled_wo = scaler.fit_transform(demo_selected_wo)

#Elbow Method to find optimal k
wcss_wo = []
k_range_wo = range(1, 11)  

for k in k_range_wo:
    kmeans_wo = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans_wo.fit(demo_scaled_wo)
    wcss_wo.append(kmeans_wo.inertia_)  #Storing WCSS values

#Plotting the Elbow Method graph
plt.figure(figsize=(8, 5))
plt.plot(k_range_wo, wcss_wo, marker="o", linestyle="--")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("WCSS (Within-Cluster Sum of Squares)")
plt.title("Elbow Method for Optimal k Selection")
plt.xticks(k_range_wo)
plt.grid(False)
plt.show()

#Printing WCSS values
for k, value in zip(k_range_wo, wcss_wo):
    print(f"k={k}, WCSS={value:.2f}")
```

### Multiple Linear Regression

To explore which variables significantly influence **EV Adoption Rate (%)**

```python
#Independent varaibles: demographic and infrastructure factors 

X = final_demographic_wo_sj[[
    'Median Household Income', 'Total Population', 'Number of Stations',
    'Charging Station Density', 'Total Registered Vehicles', 'Average EV Age',
    'Mean Commute Travel (min)', 'Education (%)', 'Median Age',
    'Average Credit Score Ranking'
]]

#Dependent variable: EV Adoption Rate (%)

y = final_demographic_wo_sj['EV Adoption Rate (%)']
```

```python
import statsmodels.api as sm

#Adding a constant term to the model
X = sm.add_constant(X)  

#Fitting the model
model = sm.OLS(y, X).fit()

#Regression results
print(model.summary())
```

**Interpretation:**

The F-Statistic indicates that the model as a whole shows it is statitically significant

Statistically significant variables include median household income, number of stations, education, median age, and average credit score rating. This means that:

- Higher income =  higher adoption
- More infraestructure = more adoption
- Slightly older counties adopt more EVs
- Financial trustworthiness might correlate with EV affordability or loan access

### Visualizing results for both the regression and clustering analysis

Variables influencing EV Adoption Rate the most

```python
#Extracting coefficients and p-values
coeffs = model.params
p_values = model.pvalues

#Create DataFrame
results_df = pd.DataFrame({'coef': coeffs, 'p_value': p_values})
results_df = results_df.drop('const')  #Excluding intercept

#Add stars for significance
results_df['significance'] = results_df['p_value'].apply(
    lambda p: '**' if p < 0.01 else '*' if p < 0.05 else '')

#Sorting by coefficient magnitude
results_df = results_df.sort_values('coef', key=abs, ascending=False)

#Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x='coef', y=results_df.index, data=results_df, palette='YlOrRd')
for i, (p, star) in enumerate(zip(results_df['p_value'], results_df['significance'])):
    plt.text(results_df['coef'].iloc[i], i, f' {star}', va='center', fontsize=12)

plt.axvline(0, color='black', linestyle='--')
plt.title('Effect of Variables on EV Adoption Rate (%)\n(p < 0.05 = *, p < 0.01 = **)')
plt.xlabel('Coefficient')
plt.ylabel('Variables')
plt.tight_layout()
plt.show()
```
